```python
import random

# Function to optimize
def func(x):
    return x**2

# Initial population
population = [random.randint(0, 31) for _ in range(10)]

# Number of generations
num_generations = 100

for generation in range(num_generations):
    print(f"Generation {generation}, Best solution {max(population, key=func)}")

    # Apply selection
    population = sorted(population, key=func, reverse=True)[:5]

    # Apply crossover
    next_generation = []
    for _ in range(5):
        parent1 = random.choice(population)
        parent2 = random.choice(population)
        child = parent1 // 2 + parent2 // 2
        next_generation.append(child)

    # Apply mutation
    population = [individual + random.choice([-1, 0, 1]) for individual in next_generation]

print(f"Final solution: {max(population, key=func)}")
```
