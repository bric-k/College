```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import skew, kurtosis
from scipy.stats import shapiro
```

```python
scores = [50,69,99,87,89,93,95,97,99,100]

mean = np.mean(scores)
median = np.median(scores)
variance = np.var(scores)
std_dev = np.std(scores)

print("Mean: ", mean)
print("Median: ", median)
print("Variance: ", variance)
print("Standard Deviation: ", std_dev)

# plotting the bell curve
plt.hist(scores, bins=10, density=True, alpha=0.6, color='g')

# Calculate the probability density function
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

# plot the pdf
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: Mean = %.2f,  Std Dev = %.2f" % (mean, std_dev)
plt.title(title)

plt.show()
```

```python

price_data = [13,43,54,34,40,56,34,61,34,23]
price_data.sort()

#Range
range_data = np.ptp(price_data)

# q1 25th%ile
q1 = np.percentile(price_data, 25)

# q3 75th%ile
q3 = np.percentile(price_data, 75)

# iqr
iqr = q3 - q1

print("Range: ", range_data)
print("25th Percentile: ", q1)
print("Interquartile Range: ", iqr)
```

```python
interest_rates = [12.05, 13, 11, 18, 10, 11.5, 15.08, 21, 6, 8, 13.2, 7.5]

#Skewness
skewness = skew(interest_rates)

#Kurtosis
kurt = kurtosis(interest_rates)

print("Skewness: ", skewness)
print("Kurtosis: ", kurt)
```

```python
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.8691]

#Shapiro-Wilk test
stat, p = shapiro(data)

print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian')
else:
    print('Sample does not look Gaussian')
```
